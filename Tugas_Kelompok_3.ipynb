```python
#16. Menampilkan top 10 negara dengan pertandingan terbanyak
country_MatchesPlayed = sport.groupby('Country')[['MatchesPlayed']].sum()
Sorted_MatchesPlayed = country_MatchesPlayed.sort_values('MatchesPlayed', ascending = False) [:15]
Sorted_MatchesPlayed
```

```python
#22. Mengklasifikasikan data worldcup berdasarkan gol [Excellent, good, dan Average]
def GoalsScored(GoalsScored):
  if GoalsScored>=80.0:
    return "Excellent"
  elif GoalsScored>=70.0:
    return "Good"
  else:
    return "Average"

sport['rate']=sport['GoalsScored'].apply(GoalsScored)
sport.head()
```
